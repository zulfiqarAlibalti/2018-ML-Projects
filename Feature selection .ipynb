#### 1.Univariate Selcetion

In this kind of feature selection by statistical test we select those feature which has srong relationship with output variable.

Here we user scikit-learn library **SelectBest** that have different statistical tests.One the that test we using here is **chi_squared(chi^^2)**

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
data = pd.read_csv(r'C:\Users\Zulfiqar Ali\Desktop\sir\excel\mobile-price-classification\train.csv')
```

```python
data.head()
data.round(0)
```

```python
X = data.iloc[: ,0:20].values
y = data.iloc[: ,-1].values  
```

```python
#Apply SelectionKBest class to extract top 60 best features
```

```python
bestfeatures = SelectKBest(score_func=chi2 ,k=10)
fit = bestfeatures.fit(X ,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns_)
#contact two dataframes for better visaulization
featureScores= pd.concat([dfcolumns ,dfscores] ,axis=1).values
featureScores.columns =['features' ,'Score'] #naming the Dataframe
print(featuresScores.nlargest(10 ,'Score')) #print 60 best features
```
